```python
import processing as ps

one_hot = True
train_data, test_data, pred_data = ps.preprocess(one__hot=one_hot)
```

```python
from activations import ReLu, Sigmoid
from costs import CrossEntropyCost, QuadraticCost, CategoricalCrossEntropyCost

from classifier import TrainingArguments, Classifier
```

```python
# Setting up training arguments
Args = TrainingArguments(
    training_data=train_data,
    epochs=100,
    mini_batch_size=128,
    learning_rate=4e-3,
    lmbda=0.3,
    test_data=test_data,
    monitor_evaluation_accuracy=True,
    monitor_evaluation_cost=False,
    monitor_training_accuracy=True,
    monitor_training_cost=False,
    optimizer="Adam"  
)
```

```python
# I would advise you to choose the CategoricalCross-Entropy if one_hot particularly when activation=ReLu
if one_hot: model = Classifier([42, 84, 3], cost=QuadraticCost, activation=Sigmoid, with_biases=True)
else: model = Classifier([16, 160, 15, 3], cost=QuadraticCost, activation=Sigmoid, with_biases=True)
```

```python
# Training the model
model.train(Args)
```

```python
# If you want to check training and test stats!
print(model.training_accuracy)
print(model.evaluation_accuracy)
print(model.training_cost)
print(model.evaluation_cost)
```

```python
# Predicting with our model
model.predict(predict_data=pred_data)
```

```python
# Saving this model with pickle
model.save(model_name="Billy")
```
